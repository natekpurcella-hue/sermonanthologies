```
import subprocess
import sys

print("Cloning repository...")
subprocess.run(['git', 'clone', 'https://github.com/nathanpurcella/sermons.git'], check=True)

print("Installing dependencies from requirements file...")
subprocess.run([sys.executable, '-m', 'pip', 'install', '-r', 'sermons/requirements_no_pyaudio.txt'], check=True)

print("Running the generation script...")
# We expect this to fail if the seed file is not present
# This is just to confirm the environment is working
try:
    subprocess.run([sys.executable, 'sermons/generate_spurgeon.py'], check=True)
except subprocess.CalledProcessError as e:
    print(f"Script failed as expected (likely missing seed file): {e}")

print("✅ Script execution finished.")
```
